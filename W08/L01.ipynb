```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import category_encoders as ca 
```

```python
new_col_name = [
    col.replace(" ", "_").lower()
    for col in df.columns
]
df.columns = new_col_name
```

```python
drop_col_list = ['id', 'license_id', 'ssa', 'location', 'application_created_date', 'account_number', 'address']
df = df.drop(drop_col_list, axis=1)  # axis 1 sutunlar axis 0 satirlar
```

```python
df['license_status_change_date'] = np.where(df.license_status_change_date.isnull(), 1,  0)  # 1 ve 0 olarak replace edildi
```

```python
pd.crosstab(df.license_status_change_date, df.license_status)  # crosstab karsilastirma komutu ile bakalim dedik
```

```python
# Zaman cizelgesi olusturma
# encoding
# feature engineering  adimlarini yapacagiz
```

```python
df.info()
```

```python
df.application_requirements_complete = pd.DatetimeIndex(df.application_requirements_complete)  # datetime cevirimleri
```

```python
df.payment_date = pd.DatetimeIndex(df.payment_date)
df.license_term_start_date = pd.DatetimeIndex(df.license_term_start_date)
```

```python
df.license_term_expiration_date = pd.DatetimeIndex(df.license_term_expiration_date)
```

```python
df.license_approved_for_issuance = pd.DatetimeIndex(df.license_approved_for_issuance)
df.date_issued = pd.DatetimeIndex(df.date_issued)
```

```python
df.info()
```

```python
# bu bolum feature engineering kismi

# Farkli basvuru tarihleri arasindaki gunleri bulacagiz
df['completion_to_start'] = (df.license_term_start_date - df.application_requirements_complete).dt.days  # yeni bir sutun yaratiyoruz
```

```python
df['start_to_expiry'] = (df.license_term_start_date - df.license_term_expiration_date).dt.days
```

```python
df['approval_to_issuance'] = (df.date_issued - df.license_approved_for_issuance).dt.days
```

```python
df.license_status
```

```python
df['Target'] = df[['license_status']].apply(lambda col:pd.Categorical(col).codes)
```

```python
df.Target.unique()
```

```python
df.isnull().sum() 
```

```python
df[['ward', 'precinct', 'ward_precinct', 'police_district']]
```

```python
df['presence_of_enquiry_details'] = np.where(df.ward.isnull() | df.precinct.isnull() | df.ward_precinct.isnull() | df.police_district.isnull(), 0, 1)
```

```python
enc = ca.OneHotEncoder().fit(df.Target.astype(str))
y_onehot = enc.transform(df.Target.astype(str))
```

```python
y_onehot.columns
```

```python
y_onehot
```

```python
class_names = y_onehot.columns
for class_ in class_names:
    enc = ca.TargetEncoder(smoothing=0)
    temp = enc.fit_transform(df[['license_description', 'state', 'city']],y_onehot[class_])
    temp.columns = [str(x)+'_'+str(class_) for x in temp.columns]
    df = pd.concat([df, temp], axis = 1)
```

```python
#under sampling
df_0 = df[df.Target==0].sample(frac=0.3, replace=False)
df_1 = df[df.Target==1].sample(frac=0.3, replace=False)

#over sampling
df_2 = df[df.Target==2].sample(frac=200, replace=True)
df_3 = df[df.Target==3].sample(frac=100, replace=True)
df_4 = df[df.Target==4].sample(frac=2, replace=True)

sample_df = pd.concat([df_0, df_1, df_2, df_3, df_4])

sample_df.Target.value_counts(normalize=True)
```
